```python
# Functions to open image, save image and get pixels from the image

from PIL import Image

# Open an Image
def open_image(path):
  newImage = Image.open(path)
  return newImage

# Save Image
def save_image(image, path):
  image.save(path, 'png')


# Create a new image
def create_image(i, j):
  image = Image.new("RGB", (i, j), "white")
  return image


# Pixel. information
def get_pixel(image, i, j):
  # Inside image bounds?
  width, height = image.size
  if i > width or j > height:
    return None

  # Get Pixel
  pixel = image.getpixel((i, j))
  return pixel
```

### 2. Red - Green - Blue pixel combination

```python
def Pixel_Change(image):
  # Get size
  width, height = image.size

  # Create new Image and a Pixel Map
  new = create_image(width, height)
  pixels = new.load()

  # Transform to grayscale
  for i in range(width):
    for j in range(height):
      # Get Pixel
      pixel = get_pixel(image, i, j)

      # Get R, G, B values (This are int from 0 to 255)
      red =   pixel[0]
      green = pixel[1]
      blue =  pixel[2]


      # Set Pixel in new image
      pixels[i, j] = (50, 50, 55)  # Refer the video presentation to het the color pixel combinations

  # Return new image
  return new

original = open_image('GL-Logo.png')
new = Pixel_Change(original)
save_image(new, 'GL-Logo_Transformed.png')
Image.open('GL-Logo_Transformed.png')
```

```python
# Gray scale image

from PIL import Image

# Open an Image
def open_image(path):
  newImage = Image.open(path)
  return newImage

# Save Image
def save_image(image, path):
  image.save(path, 'png')


# Create a Grayscale version of the image
def convert_grayscale(image):
  # Get size
  width, height = image.size

  # Create new Image and a Pixel Map
  new = create_image(width, height)
  pixels = new.load()

  # Transform to grayscale
  for i in range(width):
    for j in range(height):
      # Get Pixel
      pixel = get_pixel(image, i, j)

      # Get R, G, B pixel values
      red =   pixel[0]
      green = pixel[1]
      blue =  pixel[2]

      # Transform to grayscale
      gray = (red * 0.7) + (green * 0.7) + (blue * 0.7)

      # Set Pixel in new image
      pixels[i, j] = (int(gray), int(gray), int(gray))

  # Return new image
  return new

original = open_image('GL-Logo.png')
new = convert_grayscale(original)
save_image(new, 'GL-Logo_Transformed.png')
Image.open('GL-Logo_Transformed.png')
```

```python
# R-G-B color image

def convert_grayscale(image):
  # Get size
  width, height = image.size

  # Create new Image and a Pixel Map
  new = create_image(width, height)
  pixels = new.load()

  # Transform to grayscale
  for i in range(width):
    for j in range(height):
      # Get Pixel
      pixel = get_pixel(image, i, j)

      # Get R, G, B values (This are int from 0 to 255)
      red =   pixel[0]
      green = pixel[1]
      blue =  pixel[2]

      # Transform to primary  (change the threshold vallues to get different colors)
      if red > 150:
        red = 255
      else:
        red = 0
      if green > 150:
        green = 250
      else:
        green = 0
      if blue > 150:
        blue = 250
      else:
        blue = 0

      # Set Pixel in new image
      pixels[i, j] = (int(red), int(green), int(blue))

  # Return new image
  return new

original = open_image('GL-Logo.png')
new = convert_grayscale(original)
save_image(new, 'GL-Logo_Transformed.png')
Image.open('GL-Logo_Transformed.png')
```
