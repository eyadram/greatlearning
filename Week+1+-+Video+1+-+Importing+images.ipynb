### 1. Import different images

PIL: Python Imaging Library is a free and open-source additional library for the Python programming language that adds support for opening, manipulating, and saving many different image file formats.

```python
from PIL import Image 
  
Image_name = "1.gif"   # Use all the different types of images available on the folder
with Image.open(Image_name) as image: 
    width, height = image.size 

print("Image width:",width)    

print("Image width:",height)
```

### 3. Import image folder

CV2: OpenCV is a library of programming functions mainly aimed at real-time computer vision. Originally developed by Intel, it was later supported by Willow Garage then Itseez. The library is cross-platform and free for use under the open-source BSD license.

```python
# pip install OpenCV
import cv2
import os
import glob  # The glob module finds all the pathnames matching 

Image_Directory = "Images/" # Macbook : Enter Directory of all images 

# img_dir = "give whole path" # Windows : Enter Directory of all images 


Path = os.path.join(Image_Directory,'*g') # '*' is used to extract all files in the directory. 'g': Get Path.
Diretory_Files = glob.glob(Path)
Data = []  # EMPTY LIST
for file in Diretory_Files:
    img = cv2.imread(file)
    Data.append(img)
    
Directory_List = next(os.walk(Image_Directory))[2]
Directory_List
```

### 5. GUI to browse and display image

Tkinter: Tkinter is a Python binding to the Tk GUI toolkit. It is the standard Python interface to the Tk GUI toolkit, and is Python's de facto standard GUI.

```python
# Library
from tkinter import *
from PIL import ImageTk, Image
from tkinter import filedialog # Browse dialog box 

# Function - display the image
def display(): 
    # Select the Imagename  from a folder  
    x = dialog_box() 
  
    # opens the image 
    im = Image.open(x) 
      
    # Image resizing + applying a HQ down sampling filter 
    im = im.resize((250, 250), Image.ANTIALIAS) 
  
    # "PhotoImage" class: used to add image to widgets
    im = ImageTk.PhotoImage(im) 
   
    # Label 
    panel = Label(win, image = im) 
      
    # Set the image 
    panel.image = im 
    panel.grid(row = 2)

# Dialog box
def dialog_box(): 
    filename = filedialog.askopenfilename(title ='"pen') 
    return filename 

# Tkinter window 
win = Tk() 
win.title("Great Learning-Computer Vision-Intro to images") # GUI window title
win.geometry("640x480") # GUI image resolution

# Browse button
Browse_button = Button(win, text ='Browse Image', command = display).grid(row = 1, columnspan = 4) 
win.mainloop() 
```
